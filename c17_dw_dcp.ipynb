#### 4. Weight and Height, W/F, SM and IR Columns: convert to integers



#### Height column to int

```python
dfNew['Height'].unique()
```

```python
# Define a function to convert feet and inches to centimeters
def convert_to_cm(value):
    if isinstance(value, str):
        if "'" in value and '"' in value:
            feet, inches = value.split("'")  # Split feet and inches
            inches = int(inches[:-1])  # Remove the inch symbol
        elif "'" in value:
            feet = int(value[:-1])
            inches = 0
        elif '"' in value:
            feet = 0
            inches = int(value[:-1])
        else:
            return value  # Return the value as is if it doesn't match the expected format
        cm = (int(feet) * 12 + int(inches)) * 2.54  # Convert feet and inches to centimeters
        return str(int(cm)) + "cm"  # Convert cm to string and concatenate with "cm" suffix
    return value

# Apply the conversion function to the "Height" column
dfNew['Height'] = dfNew['Height'].apply(convert_to_cm)

# Print the updated DataFrame
print(dfNew['Height'])
```

```python
# Define a function to convert values with suffixes (e.g., 'M' for million, 'K' for thousand)
def convert_to_thousand(value):
    if isinstance(value, str):
        if 'M' in value:
            million = re.findall(r'\d+\.*\d*', value)  # Extract the numeric value (including decimal point)
            if million:
                million = sum(float(num) for num in million)  # Calculate the sum of all numeric values
                return int(million * 1000000)  # Convert million to thousand and return as an integer
        elif 'K' in value:
            thousand = re.findall(r'\d+\.*\d*', value)  # Extract the numeric value (including decimal point)
            if thousand:
                thousand = sum(float(num) for num in thousand)  # Calculate the sum of all numeric values
                return int(thousand * 1000)  # Convert thousand to thousand and return as an integer
        else:
            return value
    return value

# Apply the conversion function to the "Value" column
dfNew['Value'] = dfNew['Value'].apply(convert_to_thousand)

```
